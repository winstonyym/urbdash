```python
import pandas as pd
import json
```

```python
city = 'paris'
df = pd.read_csv(f'../data/{city}_graph/{city}_subzone_attr.csv')
colnames = ['Planning Area'] + list(df.columns[1:])
df.columns = colnames
```

```python
with open(f"../data/urbanity_indicators.json", 'r') as f:
  indicator_list = json.load(f)
```

```python
df
```

```python
rename_dict = {}
for i in indicator_list.keys():
    rename_dict[i] = ''
```

```python
rename_dict['Footprint Proportion'] = 'Building Footprint (Proportion)'
rename_dict['Footprint Mean'] = 'Mean Building Footprint (m2)'
rename_dict['Footprint Stdev'] = 'Building Footprint St.dev (m2)'
rename_dict['Perimeter Total'] = 'Total Building Perimeter (m)'
rename_dict['Perimeter Mean'] = 'Mean Building Perimeter (m)'
rename_dict['Perimeter Stdev'] = 'Building Complexity St.dev'
rename_dict['Complexity Mean'] = 'Mean Building Complexity'
rename_dict['Complexity Stdev'] = 'Building Complexity St.dev'
rename_dict['Counts'] = 'Counts'
rename_dict['PopSum'] = 'PopSum'
rename_dict['Men'] = 'Men'
rename_dict['Women'] = 'Women'
rename_dict['Elderly'] = 'Elderly'
rename_dict['Youth'] = 'Youth'
rename_dict['Children'] = 'Children'
rename_dict['Civic'] = 'Civic'
rename_dict['Commercial'] = 'Commercial'
rename_dict['Entertainment'] = 'Entertainment'
rename_dict['Food'] = 'Food'
rename_dict['Healthcare'] = 'Healthcare'
rename_dict['Institutional'] = 'Institutional'
rename_dict['Recreational'] = 'Recreational'
rename_dict['Social'] = 'Social'
rename_dict['Green View'] = 'Green View'
rename_dict['Sky View'] = 'Sky View'
rename_dict['Building View'] = 'Building View'
rename_dict['Road View'] = 'Road View'
```

```python
invert_rename_dict = {}
for k,v in rename_dict.items():
    invert_rename_dict[v] = k
```

```python
del invert_rename_dict['Counts']
del invert_rename_dict['Green View']
del invert_rename_dict['Sky View']
del invert_rename_dict['Building View']
del invert_rename_dict['Road View']
```

```python
df = df.rename(columns = invert_rename_dict)
```

```python
chosen_cols = list(invert_rename_dict.values())
```

```python
df = df.loc[:,['Planning Area']+chosen_cols]
```

```python
results = df.to_json(orient='records')
```

```python
parsed = json.loads(results)
```

```python
with open(f'../data/{city}_graph/{city}_subzone.json', 'w') as outfile:
    json.dump(parsed, outfile)
```

```python
name_to_prop = {}
for i in parsed:
    name_to_prop[i['Planning Area']] = {k.replace(' ', ''):v for k,v in i.items()}
```

```python
with open(f"../data/{city}_graph/{city}_subzone_poly.geojson", 'r') as f:
  my_geo = json.load(f)
```

```python
next(iter(my_geo['features']))['properties']
```

```python
for i in my_geo['features']:
    i['properties'] = name_to_prop[i['properties']['AREA']]
```

```python
with open(f'../data/{city}_graph/{city}_subzone_prop.geojson', 'w') as outfile:
    json.dump(my_geo, outfile)
```

```python
color_scaler = {}
for k,v in zip(df.columns[1:].values, df.max()[1:].values/255):
    color_scaler[k.replace(" ","")] = v
```

```python
with open(f'../data/{city}_graph/{city}_color_scaler.json', 'w') as outfile:
    json.dump(color_scaler, outfile)
```

```python
color_scaler
```

```python
scale_scaler = {}
for k,v in zip(df.columns[1:].values, df.median()[1:].values):
    scale_scaler[k.replace(" ","")] = v
```

```python
with open(f'../data/{city}_graph/{city}_scale_scaler.json', 'w') as outfile:
    json.dump(scale_scaler, outfile)
```
