```python
import pandas as pd
import json
```

```python
df = pd.read_csv('../data/bangkok_attributes.csv')
```

```python
results = df.to_json(orient='records')
```

```python
parsed = json.loads(results)
```

```python
with open('../data/bangkok.json', 'w') as outfile:
    json.dump(parsed, outfile)
```

```python
colnames = ['Planning Area'] + list(df.columns[1:])
```

```python
df.columns = colnames
```

```python
mapping_names = {}
for i in colnames:
    mapping_names[i] = i 
```

```python
mapping_names
```

```python
with open('urbanity_indicators.json', 'w') as outfile:
    json.dump(mapping_names, outfile)
```
